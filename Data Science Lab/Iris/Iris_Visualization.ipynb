# Loading Dataset

```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/Data Science Lab/Iris/iris.csv')
df.head()
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
```

*   **Histogram**



```python
df.hist(edgecolor='black', linewidth=1)
plt.show()
```

* The highest Sepal length ranges between 5.4 and 5.8
* The highest sepal width ranges between 3 and 3.3
* The highest Petal length ranges between 0.5 and 1.8.
* The hihest Petal width ranges between 0.2 and 0.5

*  **Boxplot**

```python
df.boxplot(by = 'variety', figsize = (16,14))
```

* Petal length,petal width and sepal length are smaller for Setosa and larger for Virginica.
* For petal length, Setosa and Versicolor has outliers.
* For petal width, Setosa has Outlier
* For sepal length, Virginica has outlier
* For sepal width, Setosa and Virginica has outlier.

*  **Scatterplot**

```python
sns.scatterplot(x='sepal.length', y='sepal.width', data=df,hue='variety')
```

* Flower with large sepal width and small sepal length is identified as Setosa.

```python
sns.scatterplot(x='sepal.length', y='petal.length', data=df,hue='variety')
```

* Flowers with petal length between 0 and 2 and sepal length between 4 and 6 is
  identified as Setosa.
* Flowers with petal length between 3 and 5 and sepal length between 4.7 and 7
  is identified as Versicolor.
* Flowers with petal length between 4.8 and 7 and sepal length between 5.5 and 8
  is identified as Virginica.


```python
sns.scatterplot(x='sepal.length', y='petal.width', data=df,hue='variety')
```

* Flowers with lower petal width and lower sepal length are Setosa.
* Flowers with higher petal width and higher sepal length are Virginica.

```python
sns.scatterplot(x='sepal.width', y='petal.length', data=df,hue='variety')
```

* Flowers with lower petal length and higher sepal width are Setosa.

```python
sns.scatterplot(x='sepal.length', y='petal.width', data=df,hue='variety')
```

* Lower sepal length and lower petal width are Setosa.
* Higher petal width are Virginica.

```python
sns.scatterplot(x='petal.length', y='petal.width', data=df,hue='variety')
```

* Lower petal width and lower petal length are Setosa.
* Higher petal width and Higher petal length are Virginica.
* In between there is Versicolor.


* **Scatter Multiple**

*  **Scatter Matrix**

*  **Bubble Chart**

```python
import numpy as np
# create data
x = df['petal.length']
y = df['petal.width']
colors = np.random.rand(150)

# use the scatter function
plt.scatter(x, y, s=150,c=colors)
plt.xlabel('petal.length')
plt.ylabel('petal.width')
plt.show()


```

* **Density Chart**

```python
sns.kdeplot(df, shade=True)
```

* **Parallel Chart**

```python
from pandas.plotting import parallel_coordinates

plt.figure(figsize=(12, 6))
parallel_coordinates(df, class_column='variety', color=['#556270', '#4ECDC4', '#C7F464'])
plt.title('Parallel Coordinates Plot of Iris Dataset')
plt.xlabel('Features')
plt.ylabel('Value')
plt.show()
```

* Setosa can easily distinguished from the other two.
* Sepal width for the three classes are overlapping.
* Petal width and petal length has a clear separation between the three classes.

* **Andrews Curve**

```python
from pandas.plotting import andrews_curves
andrews_curves(df, "variety",colormap='rainbow')
plt.show()
```

* Setosa can be distinguished easily compared to the other classes.
