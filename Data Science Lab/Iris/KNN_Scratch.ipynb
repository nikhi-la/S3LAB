# KNN scratch

```python
import numpy as np
from collections import Counter

def knn_scratch(X_train, y_train, input, k):
    distance = []
    for datapoint in X_train:
      dist=np.sqrt(np.sum((np.array(datapoint)-np.array(input))**2))  #euclidean distance
      distance.append(dist)

    distance = np.argsort(distance)[:k]  #sorting distance in ascending order and taking the index of the first k distances
    #print(distance)
    labels = [y_train[i] for i in distance]  #finding the class labels of the corresponding indices
    #print(labels)

    counter = Counter(labels)      #Counting the occurence of each class labels

    mode_value = counter.most_common(1)[0][0]  #Finding the majority class label

    return mode_value       #returning the majority class label as the final class label


knn_scratch(X_train, y_train, input, k)
```
