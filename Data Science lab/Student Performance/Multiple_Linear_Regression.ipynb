```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/Data Science lab/Student Performance/Student_Performance.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Extracurricular Activities'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Extracurricular Activities']=le.fit_transform(df['Extracurricular Activities'])
df.head()
```

```python
from sklearn.model_selection import train_test_split

x=df.drop('Performance Index',axis=1)
y=df['Performance Index']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

regressor= LinearRegression()
regressor.fit(x_train, y_train)
y_pred= regressor.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
slope = regressor.coef_
print(f'The value of slope is {slope}')
```

```python
intercept = regressor.intercept_
print(f'The value of intercept is {intercept}')
```

```python
input1 = pd.DataFrame([[7,99,1,9,1]], columns=['Hours Studied','Previous Scores','Extracurricular Activities','Sleep Hours','Sample Question Papers Practiced'])
performance=regressor.predict(input1)[0]
print(f' Performance : {performance:.2f}')
```

```python
columns = ['Hours Studied', 'Previous Scores', 'Extracurricular Activities', 'Sleep Hours', 'Sample Question Papers Practiced']

# Prompt user for input
inputdata = []
for column in columns:
  value = int(input(f"Enter value for {column}: "))
  inputdata.append(value)

inputdata=pd.DataFrame([inputdata], columns=['Hours Studied','Previous Scores','Extracurricular Activities','Sleep Hours','Sample Question Papers Practiced'])
performance=regressor.predict(inputdata)[0]
print(f'Performance : {performance:.2f}')
```
